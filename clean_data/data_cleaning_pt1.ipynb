## Circuit Dataset Cleaning

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import numpy as np
```

```python
circuits = pd.read_csv('../datasets/circuits.csv')
circuits.head()
```

```python
circuits_renamed_cols = {'name': 'circuitName', 'location': 'circuitLocation', 'country':'circuitCountry'}
circuits_drop_cols = ['lat', 'lng', 'alt', 'url']
circuits.rename(columns=circuits_renamed_cols, inplace=True)
circuits = circuits.drop(columns=circuits_drop_cols, axis=1)
circuits.columns
circuits.head()
```

## Pitstops Datset Cleaning
